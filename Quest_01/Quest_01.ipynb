# Part 1

```python
file_name = "everybody_codes_e2025_q01_p1.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

names = [i for i in data[0].split(",")]
instructions = [i for i in data[2].split(",")]

cur_pos = 0

for instruction in instructions:
    step = -int(instruction[1:]) if instruction[0] == "L"  else int(instruction[1:])
    if (cur_pos + int(step)) > len(names) - 1:
        cur_pos = len(names)- 1
    elif cur_pos + int(step) < 0:
        cur_pos = 0
    else:
        cur_pos = cur_pos +  int(step)

print(names[cur_pos])
```
