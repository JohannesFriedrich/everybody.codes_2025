# Part 1

```python
from math import modf
file_name = "everybody_codes_e2025_q02_p1.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

A = (25,9)

def add(X,Y):
    return (X[0]+Y[0], X[1]+Y[1])

def mul(X,Y):
    return (X[0] * Y[0] - X[1] * Y[1], X[0] * Y[1] + X[1] * Y[0])

assert mul((-2,5), (10,-1)) == (-15,52)

def div(X,Y):
    _ ,x  = modf(X[0]/Y[0])
    _, y= modf(X[1]/Y[1])
             
    return (int(x),int(y))

assert div((10,12), (2,2)) == (5,6)
assert div((-11,-12), (3,5)) == (-3,-2)

```

```python
R = (0,0)
def run_cycle(A, R):
    R = mul(R, R)
    R = div(R, (10,10))
    R = add(R, A)
    return R

A=(156,54)
for _ in range(0,3):
    R = run_cycle(A ,R)
print(R)

```

# Part 2

```python
A=(-79765,15616)

def run_cycle(A,R):
    R = mul(R, R)
    R = div(R, (100000,100000))
    R = add(R, A)
    return R

not_engraved = 0
grids = 0
for x in range(A[0], A[0] + 1001, 10):
    for y in range(A[1], A[1] + 1001 ,10):
        grids += 1
        R = (0,0)
        for i in range(100):
            R = run_cycle((x,y),R)
            if R[0] > 1000000 or R[0] < -1000000 or R[1] > 1000000 or R[1] < -1000000:
                not_engraved += 1
                break

grids - not_engraved
```
