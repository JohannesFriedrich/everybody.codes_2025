# Part 1

```python
file_name = "everybody_codes_e2025_q05_p1.txt"
file = open(file_name, "r")
data = file.read().strip()#.split("\n")
# data = "58:5,3,7,8,9,10,4,5,7,8,8"

id, data = data.split(":")
data = [int(i) for i in data.split(",")]

def create_fishbone(list, is_part_3 = False):
    fishbone = []
    segment = [None, list[0], None]
    fishbone.append(segment)

    for num in list[1:]:
        cur_seg = 0
        while True:
            if num > fishbone[cur_seg][1] and fishbone[cur_seg][2] is None:
                fishbone[cur_seg][2] = num
                break
            elif num < fishbone[cur_seg][1] and fishbone[cur_seg][0] is None:
                fishbone[cur_seg][0] = num
                break
            else:
                cur_seg += 1

            if cur_seg == len(fishbone):
                segment = [None, num, None]
                fishbone.append(segment)
                break

    # calculate levels for Part 3
    levels = []
    if is_part_3:
        for x in fishbone:
            level = 0
            exp_count = 0 
            for y in reversed(x):
                if y is not None:
                    level += y*10**exp_count
                    exp_count += 1

            levels.append(level)
    return "".join([str(i[1]) for i in fishbone]), levels

create_fishbone(data)[0]
```

# Part 2

```python
file_name = "everybody_codes_e2025_q05_p2.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

res = []
for line in data:
    id, line = line.split(":")
    line = [int(i) for i in line.split(",")]

    value ,_ = create_fishbone(line)
    res.append(int(value))

max(res) - min(res)
```

# Part 3

```python
file_name = "everybody_codes_e2025_q05_p3.txt"
# file_name = "test.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

all_fishbones = {}
for line in data:
    id, line = line.split(":")
    line = [int(i) for i in line.split(",")]
    value, levels = create_fishbone(line, True)
    all_fishbones.update({id: [int(value), levels]})

# sort by priority
data=[tuple(x) for x in all_fishbones.items()] 
sorted_list = [k for k, v in sorted(data, key=lambda item: (item[1][0], item[1][1], int(item[0])), reverse=True)]

res = 0
i = 1
for e in sorted_list:
    res += int(e)*i
    i += 1
res
```
