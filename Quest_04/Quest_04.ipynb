# Part 3

```python
file_name = "everybody_codes_e2025_q04_p3.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

translation = 1
for line in data:
    if "|" in line:
        g1, g2 = line.split("|")
        translation *= int(g2)/int(g1)   

int((translation * int(data[0])/int(data[-1]) * 100)//1)

```
