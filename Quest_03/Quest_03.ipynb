# Part 3

```python
from collections import Counter

file_name = "everybody_codes_e2025_q03_p3.txt"
file = open(file_name, "r")
data = file.read().strip().split(",")
data = [int(i) for i in data]

counts = Counter(data)
counts.most_common(1)[0][1]
```
